```python
import requests
import os

base_github_url = "https://raw.githubusercontent.com/MuskanBhate/databricks/dev/dataset/"
base_volume_path = "/Volumes/retail_project/raw_data/raw/"

files = [
    "customer/customers_profile.csv",
    "customer/customers_address.csv",
    "customer/customers_contact.csv",
    "customer/customers_loyalty.csv",

    "orders/orders_main.csv",
    "orders/orders_items.csv",
    "orders/orders_channel.csv",
    "orders/orders_promo.csv",
    "orders/orders_shipping.csv",

    "products/products_catalog.csv",
    "products/products_price.csv",
    "products/products_supplier.csv",

    "inventory/inventory_stock.csv",
    "inventory/inventory_cost.csv",
    "inventory/inventory_supplier.csv",

    "payments/payments_main.csv",
    "payments/payments_amount.csv",
    "payments/payments_gateway.csv",

    "returns/returns_main.csv",
    "returns/returns_logistics.csv",
    "returns/returns_refund.csv",

    "shipments/shipments_main.csv",
    "shipments/shipments_cost.csv",
    "shipments/shipments_tracking.csv",
    "shipments/shipments_warehouse.csv",

    "store/stores_info.csv",
    "store/stores_area.csv",
    "store/stores_manager.csv"
]

for file in files:
    url = base_github_url + file
    local_path = "/tmp/" + file.split("/")[-1]

    response = requests.get(url)
    with open(local_path, "wb") as f:
        f.write(response.content)

    dbutils.fs.mkdirs(base_volume_path + file.split("/")[0])

    dbutils.fs.put(base_volume_path + file, open(local_path, 'r').read(), overwrite=True)

print("✅ All files downloaded from GitHub to Volume")
```

```python
def load_to_bronze(path, table_name):
    
    df = (spark.read
          .format("csv")
          .option("header", "true")
          .option("inferSchema", "true")
          .load(path)
         )
    
    (df.write
        .format("delta")
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(f"retail_project.retail_bronze.{table_name}")
    )
```
